### **c) P(L = 1 | R = 0, T = 1)**

$P(L=1|R=0,T=1)=P(L=1|T=1)$

De la tabla:

$P(L=1|R=0,T=1)= \color{Blue} \textbf{0.3}$ <br><br><br><br>


### **d) P(L = 1 | T = 1)**
También se obtiene directamente de la tabla P(L|T):

$P(L=1|T=1)= \color{Blue} \textbf{0.3}$ <br><br><br><br>

### **e) P(L = 1 | R = 0)**

$P(L=1|R=0) = \sum_t P(L=1|T=t)P(T=t|R=0)$

Expandimos:

$ = P(L=1|T=1)P(T=1|R=0) + P(L=1|T=0)P(T=0|R=0)$

Sustituyendo:

$=(0.3)(0.1)+(0.1)(0.9)$

$=0.03+0.09$

$P(L=1|R=0)= \color{Blue} \textbf{0.12}$ <br><br><br><br>


### **f) P(T = 1 | L = 0)**

calculamos P(L=0):

$P(L=0)=P(L=0|T=1)P(T=1)+P(L=0|T=0)P(T=0)$

Como:

$P(T=0)=1-0.17=0.83$

entonces:

$P(L=0)=(0.7)(0.17)+(0.9)(0.83)$

$=0.119+0.747$

$=0.866$

Finalmente:

$P(T=1|L=0) = \frac{(0.7)(0.17)}{0.866}$

$= \frac{0.119}{0.866} \approx 0.1374$

$P(T=1|L=0)\approx \color{Blue} \textbf{0.137}$ <br><br><br><br>
