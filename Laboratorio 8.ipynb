= P(L=1|T=1)P(T=1|R=0) + P(L=1|T=0)P(T=0|R=0)

Sustituyendo:

# **Ejercicio 3**

```python


import pandas as pd



# P(G)
P_G = {
    '+': 0.1,
    '-': 0.9
}

# P(M | G)
P_M_given_G = {
    ('+', '+'): 0.667,
    ('-', '+'): 0.333,
    ('+', '-'): 0.25,
    ('-', '-'): 0.75
}

# P(B | M)
P_B_given_M = {
    ('+', '+'): 0.4,
    ('-', '+'): 0.6,
    ('+', '-'): 0.2,
    ('-', '-'): 0.8
}

# P(C | M)
P_C_given_M = {
    ('+', '+'): 0.25,
    ('-', '+'): 0.75,
    ('+', '-'): 0.5,
    ('-', '-'): 0.5
}



results = []

for G in ['+', '-']:
    for M in ['+', '-']:
        for B in ['+', '-']:
            for C in ['+', '-']:

                prob = (
                    P_G[G]
                    * P_M_given_G[(M, G)]
                    * P_B_given_M[(B, M)]
                    * P_C_given_M[(C, M)]
                )

                results.append([G, M, B, C, prob])



df = pd.DataFrame(
    results,
    columns=['G', 'M', 'B', 'C', 'P(G,M,B,C)']
)

# Mostrar tabla
print(df)



from fractions import Fraction

print("\nTabla con fracciones aproximadas:\n")

for index, row in df.iterrows():

    frac = Fraction(row['P(G,M,B,C)']).limit_denominator()

    print(
        f"G={row['G']} "
        f"M={row['M']} "
        f"B={row['B']} "
        f"C={row['C']} "
        f"-> {frac}"
    )
```



La probabilidad conjunta se calcula utilizando:

$$
P(G,M,B,C)=P(G)\,P(M|G)\,P(B|M)\,P(C|M)
$$

---

## Cálculo

Para:

$$
G=+,\ M=+,\ B=+,\ C=+
$$

se tiene:

$$
P(+,+,+,+)
=
P(G=+)\,P(M=+|G=+)\,P(B=+|M=+)\,P(C=+|M=+)
$$

Sustituyendo valores:

$$
= (0.1)(0.667)(0.4)(0.25)
$$

$$
=0.00667
$$

# **Ejercicio 4**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

GRID_SIZE = 10

# Prior uniforme
belief = np.ones((GRID_SIZE, GRID_SIZE))
belief = belief / belief.sum()

# Leer distribución de sensores
sensor_df = pd.read_csv("Sensor_Color_Distribution.csv")

print(sensor_df.head())

# Función para obtener probabilidad
def sensor_probability(distance, observed_color):

    # Filtrar por distancia
    row = sensor_df[sensor_df["dist"] == distance]

    if len(row) == 0:
        return 1e-6

    return float(row[observed_color].values[0])

def manhattan_distance(x1, y1, x2, y2):
    return abs(x1 - x2) + abs(y1 - y2)

# Actualización Bayesiana
def update_belief(belief, evidence_position, observed_color):

    new_belief = np.zeros_like(belief)

    sensor_x, sensor_y = evidence_position

    for i in range(GRID_SIZE):
        for j in range(GRID_SIZE):

            # Distancia entre fantasma hipotético y sensor
            d = manhattan_distance(i, j, sensor_x, sensor_y)

            # Likelihood
            likelihood = sensor_probability(d, observed_color)

            # Bayes
            new_belief[i, j] = likelihood * belief[i, j]

    # Normalizar
    new_belief = new_belief / new_belief.sum()

    return new_belief
```

### **Inciso a)**

```python
# Mostrar plots colorizados

evidences = [
    ((2, 3), "R"),
    ((7, 8), "G"),
    ((5, 5), "Y")
]

current_belief = belief.copy()

step = 1

for position, color in evidences:

    current_belief = update_belief(
        current_belief,
        position,
        color
    )

    plt.figure(figsize=(6,6))

    plt.imshow(current_belief, cmap="hot")

    plt.colorbar(label="Probabilidad")

    plt.title(
        f"Posterior después de evidencia {step}\n"
        f"Sensor {position} detecta {color}"
    )

    plt.xlabel("Y")
    plt.ylabel("X")

    plt.show()

    step += 1
```

Los mapas de calor muestran cómo la distribución de probabilidad del fantasma cambia conforme se agrega nueva evidencia de los sensores. Inicialmente la probabilidad era uniforme en todas las celdas, pero después de cada observación, la probabilidad comienza a concentrarse en regiones específicas del tablero. Las zonas más claras representan posiciones con mayor probabilidad, mientras que las zonas oscuras indican posiciones poco probables o prácticamente descartadas.

A medida que se incorporan más evidencias, la incertidumbre disminuye y la distribución posterior se vuelve más precisa. En la tercera evidencia puede observarse que la probabilidad queda concentrada en muy pocas celdas, indicando que el modelo tiene una mayor certeza sobre la posible ubicación del fantasma.

### **Inciso b)**

```python
# Encontrar la posición más probable (MAP)

# Obtener celda con máxima probabilidad
max_position = np.unravel_index(
    np.argmax(current_belief),
    current_belief.shape
)

max_position = (int(max_position[0]), int(max_position[1]))
max_probability = float(current_belief[max_position])

print("Celda más probable del fantasma:")
print(max_position)

print("\nProbabilidad asociada:")
print(round(max_probability, 4))

# Heatmap
plt.figure(figsize=(6,6))

plt.imshow(current_belief, cmap="hot")

plt.colorbar(label="Probabilidad")

plt.scatter(
    max_position[1],
    max_position[0],
    marker='X',
    s=200
)

plt.title("Estimación MAP de la ubicación del fantasma")

plt.xlabel("Y")
plt.ylabel("X")

plt.show()
```

La estimación MAP identifica la celda **(0,3)** como la posición más probable del fantasma, con una probabilidad aproximada de **0.1173**. En el mapa de calor, esta posición corresponde a la región más clara y marcada con una “X”, indicando que es la ubicación más consistente con toda la evidencia proporcionada por los sensores después de realizar la actualización bayesiana.
